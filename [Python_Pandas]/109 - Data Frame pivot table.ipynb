```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df=pd.read_csv(r'd:\temp\marathon_results_2016.csv',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],index_col="Bib")
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds
(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds
(x))
```

```python
df2=df.pivot_table(index="Age",columns="M/F",values=["TotalSeconds","HalfSeconds"])
```

```python
df2.swaplevel(axis="columns").sort_index(axis="columns")
```

```python
df3=df.pivot_table(index="Country",values="City",aggfunc="count")
```

```python
df3.sort_values("City",ascending=False)
```
