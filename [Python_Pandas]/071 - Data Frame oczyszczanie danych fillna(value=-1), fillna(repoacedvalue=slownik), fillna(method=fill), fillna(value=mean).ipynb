```python
import pandas as pd
import numpy as np
```

```python
fuel=pd.read_csv(r"D:\temp\fuel.csv",low_memory=False,usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'],index_col="Vehicle ID")
fuel.head()
```

```python
    fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN
```

```python
fuel.head()
```

```python
fuel.fillna(value=-1)
```

```python
fuel.head()
```

```python
replaceValues={"Class":"---","Fuel Type":"--","Combined MPG (FT1)":-1}
```

```python
fuel.fillna(value=replaceValues).head()
```

```python
combinedMean=fuel["Combined MPG (FT1)"].mean()
```

```python
combinedMean
```

```python
fuel["Combined MPG (FT1)"].fillna(value=combinedMean,inplace=True)
```

```python
fuel["Combined MPG (FT1)"].fillna(method="ffill").head()
```
