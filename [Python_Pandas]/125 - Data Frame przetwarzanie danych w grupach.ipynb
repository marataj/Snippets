```python
import pandas as pd
import numpy as np
```

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv(r'd:\temp\marathon_results_2016.csv', index_col='Bib',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
groupCity=df.groupby(by="City")
```

```python
groupCity.head()
```

```python
for row, content in groupCity:
    print(row,"\n",content.loc[content["TotalSeconds"].idxmin()])
```

```python
theBestPerCity=pd.DataFrame()
```

```python
for row, content in groupCity:
    theBestPerCity=theBestPerCity.append(content.nsmallest(1,"TotalSeconds"))
```
